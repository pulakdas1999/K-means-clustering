```python
import matplotlib.pyplot as plt

x = [4, 5, 10, 4, 3, 11, 14 , 6, 10, 12]
y = [21, 19, 24, 17, 16, 25, 24, 22, 21, 21]

plt.scatter(x, y)
plt.show()
```

```python
from sklearn.cluster import KMeans

data = list(zip(x, y))
print(data)
inertias = []

for i in range(1,11):
    kmeans = KMeans(n_clusters=i, n_init='auto')
    kmeans.fit(data)
    inertias.append(kmeans.inertia_)

#plt.scatter(range(1,11), inertias, marker='o')
plt.plot(range(1,11), inertias, marker='o')
plt.title('Elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```

```python
kmeans = KMeans(n_clusters=3, n_init='auto')
kmeans.fit(data)
print(type(kmeans.labels_), kmeans.labels_, type(kmeans.cluster_centers_), kmeans.cluster_centers_, sep='\n')

centroids = kmeans.cluster_centers_
plt.scatter(x, y, c=kmeans.labels_)
plt.scatter(centroids[:,0] , centroids[:,1], s = 200, marker = 'X')
plt.show()
```

```python
X = [4, 5, 9, 6, 1, 3, 2, 8, 0, 7]
Y = [7, 3, 8, 1, 4, 2, 0, 9, 5, 6]
Z = [1, 4, 3, 5, 2, 8, 6, 0, 7, 9]
data = list(zip(X, Y, Z))
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
inertias = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, n_init='auto')
    kmeans.fit(data)
    inertias.append(kmeans.inertia_)

#plt.scatter(range(1,11), inertias, marker='o')
plt.plot(range(1,11), inertias, marker='o')
plt.title('Elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```

```python
kmeans = KMeans(n_clusters=3, n_init='auto')
kmeans.fit(data)
print(kmeans.labels_, kmeans.cluster_centers_, sep='\n')

centroids = kmeans.cluster_centers_
plt.scatter(centroids[:,0] , centroids[:,1], s = 200, marker = 'X')
plt.show()
```
